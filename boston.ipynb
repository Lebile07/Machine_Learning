```python
import pandas as pd
import numpy as np

data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
X = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
y = raw_df.values[1::2, 2]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X.shape
```

```python
y.shape
```

## Training Agent with Crim Features

```python
X[0]
```

```python
X_CRIM=X[:,0]
```

```python
X_CRIM=X_CRIM.reshape(-1,1)
```

```python
X_CRIM.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X_CRIM,y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
boston_model_crim =LinearRegression()
boston_model_crim.fit(X_train,y_train)
print(boston_model_crim.coef_)
print(boston_model_crim.intercept_)
```

```python
y_pred_crim=boston_model_crim.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred_crim)
mse**0.5
```

```python
r2_score(y_test,y_pred_crim)*100
```
