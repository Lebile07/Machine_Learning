```python
from sklearn.datasets import load_breast_cancer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
```

```python
bunch=load_breast_cancer()
```

```python
X=bunch.data
```

```python
y=bunch.target
```

```python
def KNN(n,p,w):
    classifier=KNeighborsClassifier(n_neighbors=n,p=powers,weights=weight)
    X_train,X_test,y_train,y_test=train_test_split(X,y)
    classifier.fit(X_train,y_train)
    y_pred=classifier.predict(X_test)
    # score = r2_score(y_test,y_pred)
    score=classifier.score(X_test,y_test)
    print(f"Score : {score}")
    # print("Score",r2_score(y_test,y_pred))
    # print("rmse",mean_squared_error(y_test,y_pred)**0.5)
    print(f"{classification_report(y_test,y_pred)}")
    print(confusion_matrix(y_test,y_pred))
    
```

## when the model use 3 neighbors, uniform weight and power parameters ranging from 1-4

```python
choice=3
for i in range(1,5):
    weight='uniform'
    powers=i
    n_neighbors= choice
    print(f"n_neighbors is {n_neighbors}")
    print(f"Weight is uniform")
    print(f"When power parameters is {i}")
    KNN(choice, powers, weight)
    print("="*55)
```
